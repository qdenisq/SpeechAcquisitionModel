## 1 - Load audio

```python
from scipy.io import wavfile as wav
from python_speech_features import mfcc
from matplotlib import pyplot as plt
import numpy as np
```

```python
fname = r'C:\Study\SpeechAcquisitionModel\src\reinforcement\vtl_reference_guided\ref.wav'
sr, input_audio = wav.read(fname)
input_audio = input_audio[:sr]
```

## 2 - Calc and Display MFCCs 

```python
num_cep=20
winlen = 0.02
out = mfcc(input_audio, 
           samplerate=sr, 
           numcep=num_cep + 1, 
           winlen=winlen, 
           nfft=int(2*sr*winlen),
           winfunc=np.hamming, 
           winstep=winlen/2)
```

```python
# print(out[:10])
plt.imshow(out.T)
plt.colorbar()
```

```python
print(out[5:10])
```

```python
mfcc_delta = np.array([out[i+1] - out[i] for i in range(out.shape[0] - 1)])
```

```python
plt.imshow(abs(mfcc_delta.T))
plt.colorbar()
plt.show()
```
